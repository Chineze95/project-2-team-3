```python
import pandas as pd
```

```python
#relevant files 

rottentomatoes = pd.read_csv('rottenTomatoes_cleaned2.csv', error_bad_lines=False)
metacritic = pd.read_csv('metaCritic_cleaned2.csv')
imdb = pd.read_csv('imdbDBJoined.csv', error_bad_lines=False)
moviedb = pd.read_csv('tmd_movies.csv')
```

```python
metacritic.head(2)
```

```python
rottentomatoes.head(2)
```

```python
merge_df = pd.merge(metacritic, rottentomatoes, on="movie_title")
```

```python
merge_df.head(2)
```

```python
imdb.head(2)
```

```python
imdb = imdb.rename(columns={'title':'movie_title'})
imdb.count()
```

```python
merge_df2 = pd.merge(merge_df, imdb, on="movie_title")
merge_df2['imdb_title_id'].values
```

```python
moviedb.head(2)

moviedb = imdb.rename(columns={'imdb_id':'imdb_title_id'})
moviedb.columns
```

```python
merge_df3 = pd.merge(merge_df2, moviedb, on="imdb_title_id")
```

```python
merge_df3
```

```python
merge_df3.to_csv('Final_Movie_DB', header=True, index=False)
```
