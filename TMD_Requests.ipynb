```python
import requests
import json
from config import api_key
import pandas as pd
import time
from pprint import pprint
```

```python
# Import Metacritic CSV file

mc_file_path = "metacritic_movies.csv"

mc_df = pd.read_csv(mc_file_path)

mc_df.head()
```

```python
# Get a list of the movie titles from the metacritic dataframe

titles = []

for index, row in mc_df.iterrows():
    titles.append(row['movie_title'])
    
titles
```

```python
# Loop through movie titles
#  - Query by title key words
#  - From the query results, get the ids of the results that exactly matches the title
#  - Then look up the movie id and build an output dictionary of certain values
#  - Append the output dicitonaries to a list


url_q = "https://api.themoviedb.org/3/search/movie?api_key=" + api_key + "&query="
url_id1 = "https://api.themoviedb.org/3/movie/"
url_id2 = "?api_key=" + api_key + "&language=en-US"

movie_list = []

i = 0
n = 0

for title in titles:
    i += 1

    lookup_string = title.replace(' ', '%20')
    #print(lookup_string)
    response_q = requests.get(url_q + lookup_string)
    result_q = response_q.json()
    
    time.sleep(1)
    
    if 'results' in result_q:

        for movie in result_q['results']:

            if movie['title'] == title:
                n += 1
                print(i, n)

                movie_id = movie['id']
                response_id = requests.get(url_id1 + str(movie_id) + url_id2)
                result_id = response_id.json()

                movie_dict = {}

                movie_dict['budget'] = result_id['budget']
                movie_dict['id'] = result_id['id']
                movie_dict['popularity'] = result_id['budget']
                movie_dict['popularity'] = result_id['popularity']
                movie_dict['imdb_id'] = result_id['imdb_id']
                movie_dict['release_date'] = result_id['release_date']
                movie_dict['runtime'] = result_id['runtime']
                movie_dict['title'] = result_id['title']
                movie_dict['vote_average'] = result_id['vote_average']
                movie_dict['vote_count'] = result_id['vote_count']

                movie_list.append(movie_dict)
            

  
```

```python
# Ingest the list of movie dictionaries into a dataframe

api_movies_df = pd.DataFrame(movie_list)

# Write the dataframe to a csv file

api_movies_df.to_csv('tmd_movies.csv', header=True, index=True)
```
